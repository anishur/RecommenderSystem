```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id":"id"})
credits_column_renamed.head()
```

```python
movie_df_merge = movies_df.merge(credits_column_renamed, on='id')
```

```python
movies_cleaned_df = movie_df_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df.info()
```

```python
# Using Weighted average for each movie´s Average Rating
# W = (Rv + Cm)/(v+m) 
# where, W = weighted Rating, R = average for the movie as a number from 0 to 10 (mean)=(Rating),
# v = average for the movie = (votes),m = minimum votes required to be listed in the Top 250 (currently 3000)
# C = the mean vote across the whole report (currently 6.9)
# source http://trailerpark.weebly.com/imdb-rating.html?source=post_page
```

```python
# Calculate all the components based on the above formula

v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
m = movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_average'] = ((R*v) + (C*m)) / (v + m)
```

```python
movies_cleaned_df.head()
```

```python
movie_sorted_ranking = movies_cleaned_df.sort_values('weighted_average', ascending=False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

# Content Based Recommendation System

```python
movies_cleaned_df.head(1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Using Abhishek Thakur´s arguments for TF-IDF

tfv = TfidfVectorizer(min_df = 3, max_features=None, strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}',ngram_range = (1,3),use_idf=1, smooth_idf=1, sublinear_tf=1, stop_words = 'english')
```

```python
# Filling NaNs with empty string
```

```python
movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna('')
```

```python
# Fitting the TF-IDE on the 'overview' text
```

```python
tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
# Compute the sigmoid kernel
```

```python
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
# Reverse mapping of indices and movie titles
```

```python
indices = pd.Series(movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
    
    # Get the index corresponding to original_title
    
    idx = indices[title]
    
    # Get the pairwise similarity scores
    
    sig_scores = list(enumerate(sig[idx]))
    
    # Sort the movies
    
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    
    # Score of the 10 most similar movies
    
    sig_scores = sig_scores[1:11]
    
    # Movie indices
    
    movie_indices = [i[0] for i in sig_scores]
    
    # Top 10 most similar movies
    
    return movies_cleaned_df['original_title'].iloc[movie_indices]
```

```python
# Testing the content-based recommendation system with the seminal film Spy Kids
```

```python
give_rec('Spy Kids')
```
